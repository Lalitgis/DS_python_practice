<h1> How to create dummy variables in Pandas </h1>

```python
import pandas as pd
```

```python
df = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
df.head()
```

<h3> First method </h3>

```python
df['Sex_male'] = df.Sex.map({'female':0, 'male':1})
```

```python
df
```

<h3> Next and frequenlty used method </h3>

```python
df.Sex.str.get_dummies()
```

```python
help(pd.get_dummies)
```

```python
# Dropping the first column
df.Sex.str.get_dummies().iloc[:,1:]
```

```python
pd.get_dummies(df.Sex, dtype = float, prefix='Sex').iloc[:,1:]
```

<h3> Next style </h3>

```python
df
```

```python
df.Embarked.value_counts()
```

```python
pd.get_dummies(df.Embarked, prefix='Embarked', dtype=float)
```

```python
embk_dummy = pd.get_dummies(df.Embarked, prefix='Embarked', dtype=float).iloc[:, 1:]
```

```python
pd.concat([df, embk_dummy], axis=1)
```

```python
df
```

```python
pd.get_dummies(df, columns=['Sex', 'Embarked'],dtype=float)
```

```python
pd.get_dummies(df, columns=['Sex', 'Embarked'],dtype=float, drop_first=True)
```
